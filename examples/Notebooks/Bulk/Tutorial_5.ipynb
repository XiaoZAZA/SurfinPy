```python
from surfinpy import utils as ut
from surfinpy import bulk_mu_vs_t as bmvt
from surfinpy import data
import numpy as np
from seaborn import palettes

import matplotlib.pyplot as plt
colors = ["#4878D0", "#EE854A", "#6ACC64", "#D65F5F", "#956CB4", "#8C613C", "#DC7EC0", "#797979", "#D5BB67", "#82C6E2"]
```

```python
temperature_range = [200, 600]
```

```python
bulk = data.ReferenceDataSet(cation = 1, anion = 1, energy = -92.0, funits = 10, file = 'bulk_vib.yaml', entropy=True, temp_range=temperature_range)
```

```python
MgO = data.DataSet(cation = 10, x = 0, y = 0, energy = -92.35637426, color=colors[0],
                   label = "Periclase", entropy = True, file = 'MgO_vib.yaml', 
                   funits = 10, temp_range=temperature_range)

Art = data.DataSet(cation = 10, x = 5, y = 20, energy = -467.308589,  color=colors[1],
                   label = "Artinite", entropy = True, file = 'Art_vib.yaml', 
                   funits = 5, temp_range=temperature_range)

Bru = data.DataSet(cation = 10, x = 0, y = 10, energy = -227.3449021,  color=colors[2],
                   label = "Brucite", entropy = True, file = 'Bru_vib.yaml', 
                   funits =  10, temp_range=temperature_range)

Nes = data.DataSet(cation = 10, x = 10, y = 30, energy = -705.7643942,  color=colors[3],
                   label = "Nesquehonite", entropy = True, file = 'Nes_vib.yaml', 
                   funits = 10, temp_range=temperature_range)

Mag = data.DataSet(cation = 10, x = 10, y = 0, energy = -310.3283693,  color=colors[4],
                   label = "Magnesite", entropy = True,  file = 'Mag_vib.yaml', 
                   funits =  10, temp_range=temperature_range)

Lan = data.DataSet(cation = 10, x = 10, y = 50, energy = -971.8116011,  color=colors[5],
                   label = "Lansfordite", entropy = True, file = 'Lan_vib.yaml', 
                   funits =  10, temp_range=temperature_range)

Hyd = data.DataSet(cation = 10, x = 8, y = 10, energy = -398.6071433,  color=colors[6],
                   label = "Hydromagnesite", entropy = True, file = 'Hyd_vib.yaml', 
                   funits =  2, temp_range=temperature_range)
```

```python
data = [MgO, Art, Bru, Nes, Lan, Hyd]
```

```python
deltaX = {'Range': [ -1, 0.6],  'Label': 'CO_2'}
deltaY = {'Range': [ 200, 600], 'Label': 'Temperature'}
x_energy=-20.53412969
y_energy=-12.83725889
mu_z = 0
```

```python
exp_x = ut.temperature_correction_range("CO2.txt", deltaY)
exp_z = ut.temperature_correction_range("H2O.txt", deltaY)
```

```python
system = bmvt.calculate(data, bulk, deltaX, deltaY, x_energy, y_energy, mu_z, exp_x, exp_z)
ax = system.plot_mu_vs_t()
```

```python
ax = system.plot_p_vs_t(temperature=273, colourmap="RdYlBu")
```

```python
ax = system.plot_mu_vs_t_vs_p(temperature=273, colourmap="RdYlBu")
```
