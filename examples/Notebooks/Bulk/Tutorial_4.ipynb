## SurfinPy

#### Tutorial 4 - Pressure vs Temperature

SurfinPy has the functionality to generate phase diagrams as a function of pressure vs temperature.

##### code

```python
from surfinpy import utils as ut
from surfinpy import bulk_mu_vs_t as bmvt
from surfinpy import data
```

```python
temperature_range = [273, 373]
```

```python
bulk = data.ReferenceDataSet(cation = 1, anion = 1, energy = -92.0, funits = 10, temp_range=temperature_range)
```

```python
MgO = data.DataSet(cation = 10, x = 0, y = 0, energy = -92.35637426, label = "Periclase", temp_range=temperature_range)
Art = data.DataSet(cation = 10, x = 5, y = 20, energy = -467.308589, label = "Artinite", temp_range=temperature_range)
Bru = data.DataSet(cation = 10, x = 0, y = 10, energy = -227.3449021, label = "Brucite", temp_range=temperature_range)
Nes = data.DataSet(cation = 10, x = 10, y = 30, energy = -705.7643942, label = "Nesquehonite", temp_range=temperature_range)
Mag = data.DataSet(cation = 10, x = 10, y = 0, energy = -310.3283693, label = "Magnesite", temp_range=temperature_range)
Lan = data.DataSet(cation = 10, x = 10, y = 50, energy = -971.8116011, label = "Lansfordite", temp_range=temperature_range)
Hyd = data.DataSet(cation = 10, x = 8, y = 10, energy = -398.6071433, label = "Hydromagnesite", temp_range=temperature_range)
```

```python
data = [MgO, Art, Bru, Nes,  Mag, Lan, Hyd]
```

```python
deltaX = {'Range': [ -1, 0.6],  'Label': 'CO2'}
deltaY = {'Range': [ 273, 373], 'Label': 'Temperature'}
x_energy=-20.53412969
y_energy=-12.83725889
mu_z = 0
```

```python
exp_x = ut.temperature_correction_range("CO2.txt", deltaY)
exp_z = ut.temperature_correction_range("H2O.txt", deltaY)
```

```python
system = bmvt.calculate(data, bulk, deltaX, deltaY, x_energy, y_energy, mu_z, exp_x, exp_z)
system.plot_mu_vs_t(set_style="seaborn-dark-palette", colourmap="RdYlBu")
```

```python
system.plot_p_vs_t(temperature=273, set_style="seaborn-dark-palette", colourmap="RdYlBu")
```

```python
system.plot_mu_vs_t_vs_p(temperature=273, set_style="seaborn-dark-palette", colourmap="RdYlBu")
```
